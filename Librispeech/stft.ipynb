```
def stft(arr,sampling_rate, plot):
  f, t, Zxx = signal.stft(arr,sampling_rate)
  #amp = 2 * np.sqrt(2)
  if plot == true :
    plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
    plt.title('STFT Magnitude')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    plt.show()
```

```
def melspectrogram(arr,sr,plot):
  arr_numpy= np.array(arr,dtype= np.float32)
  spectro= librosa.feature.melspectrogram(y=arr_numpy, sr=sr, S=None, n_fft=800, hop_length=200, power=2.0 ,n_mels=80)
 
  if plot == 1 :
    dp=librosa.power_to_db(spectro,ref=np.max),
    librosa.display.specshow(librosa.power_to_db(spectro,ref=np.max), y_axis='mel', fmax=8000, x_axis='time')
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel spectrogram')
    plt.tight_layout()
    return spectro
```
